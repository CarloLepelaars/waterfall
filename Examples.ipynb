```python
from waterfallcharts import quick_charts as qc
```

```python
a = ['sales','returns','credit fees','rebates','late charges','shipping']
b = [10,-30,-7.5,-25,95,-7]
```

```python
plot = qc.waterfall(a,b)
```

```python
plot = qc.waterfall(a,b, sorted_value=True, threshold=0.2)
```

```python
plot = qc.waterfall(a,b, sorted_value=True, threshold=0.2, formatting="$ {:,.1f}", net_label="end result", 
                    other_label= "misc", Title = 'Chart', x_lab = "", y_lab = "money", 
                   blue_color = "blue", green_color = "#95FF24", red_color = 'r')
```
